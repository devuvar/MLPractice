```python
#to make it work on both Python 2 and 3, lets import print method
from __future__ import print_function
```

```python
#lets import some modules useful for data exploration
import numpy as np

import pandas as pd
#avoid truncation of table columns
pd.set_option('display.max_columns', 100)

from matplotlib import pyplot as plt
#display plots in the notebook
%matplotlib inline

import seaborn as sns
```

```python
#load data from csv
df = pd.read_csv('real_estate_data.csv')
```

```python
df.dtypes
```

# Distribution of numerical features

```python
df.hist(figsize=(15,15))
plt.show()
```

# Distribution of categorical features

```python
for i in df.dtypes[df.dtypes=='object'].index:
    sns.countplot(y=i, data=df)
    plt.show()
```

```python
for i in df.dtypes[df.dtypes=='object'].index:
    print(df[i].value_counts())
    print()
```

# Segmetations

```python
sns.boxplot(y='property_type', x='tx_price', data=df)
plt.show()
```

```python
sns.boxplot(y='property_type', x='sqft', data=df)
plt.show()
```

```python
df.groupby('property_type').agg(['mean', 'std'])
```

# Correlations

```python
corr = df.corr()
```

```python
corr
```

```python
plt.figure(figsize=(14,12))
sns.set_style('white')
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr*100,annot=True,fmt='.0f',cbar=False, mask=mask)
plt.show()
```
