### Correlations

```python
corr = df.corr()
```

```python
corr
```

```python
plt.figure(figsize=(14,12))
sns.set_style('white')
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr*100,annot=True,fmt='.0f',cbar=False, mask=mask)
plt.show()
```

## Data Cleaning

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.basement.unique()
```

```python
df.basement = df.basement.fillna(0)
```

```python
df.basement.unique()
```

### typos and capitalization

```python
df.roof.unique()
```

```python
df.roof.replace('composition', 'Composition', inplace=True)
df.roof.replace('asphalt', 'Asphalt', inplace=True)
df.roof.replace(['shake-shingle', 'asphalt,shake-shingle'], 'Shake Shingle', inplace=True)
```

```python
df.exterior_walls.unique()
```

```python
df.exterior_walls.replace('Rock, Stone', 'Masonry', inplace=True)
df.exterior_walls.replace(['Concrete', 'Block'], 'Concrete Block', inplace=True)
```

### remove outlier

```python
sns.boxplot(df.tx_price)
plt.show()
```

```python
sns.violinplot(df.tx_price)
plt.show()
```

```python
for i in df.dtypes[df.dtypes!='object'].index:
    sns.violinplot(df[i])
    plt.show()
```

```python
#looks like a outlier in lot_size
df.lot_size.sort_values(ascending=False).head()
```

```python
df = df[df.lot_size<440000]
```

```python
df.shape
```

### Missing categorical data

```python
df.select_dtypes(include=['object']).isnull().sum()
```

```python
df.exterior_walls.fillna('Missing', inplace=True)
df.roof.fillna('Missing', inplace=True)
```

```python
df.select_dtypes(include=['object']).isnull().sum()
```
